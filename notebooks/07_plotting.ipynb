# Plotting with `Plots.jl`


In the exercise on packages, we installed the `Plots.jl` package. Now, let's see how to use it to create some basic plots.


Let us start by importing the `Plots` package:


````julia
using Plots
````

Let's create some sample data to plot. We'll create a simple sine wave.


````julia
x = 0:0.01:2Ï€
````

Let's evaluate the sine, cosine, and logarithm functions over the range of `x` values we created earlier.


````julia
y1 = sin.(x)
y2 = cos.(x)
y3 = @. log(1 + x);
````

To plot the data, we can use the `plot` function from the `Plots` package. Here's how we can use it to plot our sine wave:

```julia
plot(x_values, y_values, title="...", xlabel="...", ylabel="...", label="...")
```


````julia
plot(
    x,
    y1,
    title="Sine Function",
    xlabel="x",
    label="sin(x)",
    ylabel="y",
)
````

Say we want to scatter plot the logarithmic function on top of the existing graph, we can use `scatter!` function. Notice `!` at the end of the function, having the same meaning as before, _i.e._, it is a mutating function.


````julia
scatter!(
    x,
    y3,
    label="log(1 + x)",
    markershape=:circle,
    markersize=2,
)
````

We specify the marker shape and size using the `markershape` and `markersize` attributes, respectively. For more details about different attributes can be found [here](https://docs.juliaplots.org/dev/generated/attributes_series/).


## Subplots


We can create subplots using the `layout` attribute in the `plot` function. Here's an example of how to create a 2x2 grid of subplots:


````julia
# Define all the different plots we want to include in the subplots
p1 = plot(x, sin.(x), title="Sine Wave", xlabel="x", ylabel="sin(x)", label="sin(x)")
p2 = plot(x, cos.(x), title="Cosine Wave", xlabel="x", ylabel="cos(x)", label="cos(x)")
p3 = plot(x, log.(x), title="Logarithm", xlabel="x", ylabel="log(x)", label="log(x)")
p4 = plot(x, exp.(x), title="Exponential", xlabel="x", ylabel="exp(x)", label="exp(x)")

# Create a 2x2 grid of subplots
plot(p1, p2, p3, p4, layout=(2, 2), size=(900, 600), plot_title="Subplots Example")
````

Using the `layout` attribute, we can specify the arrangement of subplots. In this case, `(2, 2)` indicates a grid with 2 rows and 2 columns. We can also set the overall size of the figure using the `size` attribute.

To have a title for the entire figure, we have used the `plot_title` attribute.


Say we want to save the figure we created above, we can use the `savefig` function. Here's how to save the figure as a PNG file


````julia
savefig("subplots_example.png")
````
